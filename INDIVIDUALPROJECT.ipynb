```python
import pandas as pd
import matplotlib.pyplot as plt

# Data for Novak Djokovic
djokovic_data = {'Surface': ['Hard Court', 'Clay'],
                 'Wins': [5+1+1, 3],
                 'Matches': [5+1+1+1+1, 3+1]}
df_djokovic = pd.DataFrame(djokovic_data)
df_djokovic['Player'] = 'Novak Djokovic'

# Data for Jannik Sinner
sinner_data = {'Surface': ['Hard Court', 'Clay'],
               'Wins': [6+0+4+1+6+0, 3],
               'Matches': [6+0+4+0+4+1+6+0, 3+1]}
df_sinner = pd.DataFrame(sinner_data)
df_sinner['Player'] = 'Jannik Sinner'

# Data for Carlos Alcaraz
alcaraz_data = {'Surface': ['Hard Court', 'Clay'],
                'Wins': [4+1+2+1+6+0+3+1, 5+0],
                'Matches': [4+1+2+1+6+0+3+1+2, 5+0]}
df_alcaraz = pd.DataFrame(alcaraz_data)
df_alcaraz['Player'] = 'Carlos Alcaraz'

# Combine dataframes
df_combined = pd.concat([df_djokovic, df_sinner, df_alcaraz], ignore_index=True)

# Calculate win percentage
df_combined['Win Percentage'] = (df_combined['Wins'] / df_combined['Matches']) * 100

# Plotting
plt.figure(figsize=(15, 10))

# Plot Novak Djokovic
plt.subplot(2, 3, 1)
djokovic_df = df_combined[df_combined['Player'] == 'Novak Djokovic']
plt.bar(djokovic_df['Surface'], djokovic_df['Win Percentage'], color='blue')
plt.title('Novak Djokovic Win Percentage')
plt.xlabel('Surface')
plt.ylabel('Win Percentage')
plt.ylim(0, 100)  # Setting y-axis limit from 0 to 100 for percentage

# Plot Jannik Sinner
plt.subplot(2, 3, 2)
sinner_df = df_combined[df_combined['Player'] == 'Jannik Sinner']
plt.bar(sinner_df['Surface'], sinner_df['Win Percentage'], color='orange')
plt.title('Jannik Sinner Win Percentage')
plt.xlabel('Surface')
plt.ylabel('Win Percentage')
plt.ylim(0, 100)  # Setting y-axis limit from 0 to 100 for percentage

# Plot Carlos Alcaraz
plt.subplot(2, 3, 3)
alcaraz_df = df_combined[df_combined['Player'] == 'Carlos Alcaraz']
plt.bar(alcaraz_df['Surface'], alcaraz_df['Win Percentage'], color='green')
plt.title('Carlos Alcaraz Win Percentage')
plt.xlabel('Surface')
plt.ylabel('Win Percentage')
plt.ylim(0, 100)  # Setting y-axis limit from 0 to 100 for percentage

plt.tight_layout()
plt.show()
```
